```python
# Dependencies
import requests
import json
import pandas as pd

from config import gkey

```

```python
df = pd.read_csv("dc_crime_data_2019.csv")
df
```

```python
lat_lon_df = df[["LATITUDE", "LONGITUDE"]]
lat_lon_df
```

```python
# Make request for zip code from google API
zipcodes = []
for lat, lon in zip(lat_lon_df["LATITUDE"], lat_lon_df["LONGITUDE"]):
    target_url = f"https://maps.googleapis.com/maps/api/geocode/json?latlng={lat},{lon}&key={gkey}"
    geo_data = requests.get(target_url).json()
    zipcode = geo_data["results"][0]["address_components"][6]["long_name"]
    print(zipcode)
    zipcodes.append(zipcode)
print(zipcodes)
    
```
